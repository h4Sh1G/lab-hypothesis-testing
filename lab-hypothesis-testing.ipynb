```python
# standardize and remove spaces on column names

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

df
```

```python
#code here

# Null Hypothesis: Pokemon type Dragon have on average more HP than Pokemon type grass
# Alternative Hypothesis: Pokemon type Dragon have on average less HP than Pokemon type grass

alpha = 0.05

# Filter the data

df_dragon = df[df['type_1'] == 'Dragon']

df_grass = df[df['type_1'] == 'Grass']

# sample mean

mean_dragon = df_dragon['hp'].mean()

mean_grass = df_grass['hp'].mean()

# sample standard deviation

std_dragon = df_dragon['hp'].std()

std_grass = df_grass['hp'].std()

# sample size

n_dragon = len(df_dragon)

# null hypothesis

mu = mean_dragon > mean_grass

# determine p-value

stat = (mean_dragon - mu) / (std_dragon / np.sqrt(n_dragon))

stat
```

```python
p_value = st.t.sf(abs(stat), n_dragon-1)

p_value
```

```python
if p_value < alpha:
    print("We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here

# H0: There is no difference in the mean stat between Legendary and Non-Legendary PokÃ©mon.
# H1: There is a difference in the mean stat between Legendary and Non-Legendary PokÃ©mon.

# Perform t-tests

# Split the data into Legendary and Non-Legendary
legendary = df[df["legendary"] == True]
non_legendary = df[df["legendary"] == False]

stats_columns = ["hp", "attack", "defense", "sp._atk", "sp._def", "speed"]
results = []

# Perform t-tests for each stat
for stat in stats_columns:
    t_stat, p_value = st.ttest_ind(legendary[stat], non_legendary[stat], equal_var=False)
    results.append((stat, t_stat, p_value))

# Convert to DataFrame for better readability
results_df = pd.DataFrame(results, columns=["Stat", "t-statistic", "p-value"])

results_df

```

Since the p-value is less than 0.05 in both cases, we reject the null hypothesis, and we can conclude that on average the stats of legendary pokemons are different from non-legendary pokemons.

**We posit that houses close to either a school or a hospital are more expensive.**

- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

To test the hypothesis that houses close to either a school or a hospital are more expensive, we need to follow these steps:

1.	Calculate the Euclidean distance from each house to the school and hospital.
2.	Divide the dataset into two groups: houses close to either a school or hospital, and houses that are not.
3.	Perform an appropriate statistical test to compare the mean prices of the two groups.
4.	Interpret the results with a significance level of 5%.

```python
# School and Hospital coordinates
school_coords = (-118, 37)
hospital_coords = (-122, 34)

# Function to calculate Euclidean distance
def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Calculate distances to school and hospital
df['dist_to_school'] = df.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], *school_coords), axis=1)
df['dist_to_hospital'] = df.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], *hospital_coords), axis=1)

# Determine if a house is close to either the school or the hospital
df['close_to_school_or_hospital'] = (df['dist_to_school'] < 0.50) | (df['dist_to_hospital'] < 0.50)

# Divide the dataset into two groups
close_group = df[df['close_to_school_or_hospital'] == True]
far_group = df[df['close_to_school_or_hospital'] == False]

# Perform two-sample t-test
t_stat, p_value = st.ttest_ind(close_group['median_house_value'], far_group['median_house_value'], equal_var=False)

# Display the results
results = {
    "t-statistic": t_stat,
    "p-value": p_value,
    "mean_price_close": close_group['median_house_value'].mean(),
    "mean_price_far": far_group['median_house_value'].mean()
}

results_df = pd.DataFrame([results])

# Interpretation
if p_value < 0.05:
    interpretation = "Reject the null hypothesis. There is a significant difference in house prices."
else:
    interpretation = "Fail to reject the null hypothesis. There is no significant difference in house prices."

results_df["interpretation"] = interpretation
results_df
```
